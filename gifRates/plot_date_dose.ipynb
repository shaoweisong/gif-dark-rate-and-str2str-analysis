```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

file_path = "/afs/cern.ch/user/s/shsong/CMSSW_12_0_0_pre2/src/gif-dark-rate-and-str2str-analysis/gifRates/me11_txt/anode_me11_all.txt"  
data = []

with open(file_path, "r") as file:
    lines = file.readlines()

    for line in lines:
        parts = line.split()
        if len(parts) < 3 or "Processed" in line:  
            continue

        charge = parts[-2]
        date = parts[-1] 
        data.append([date, charge])

df = pd.DataFrame(data, columns=["Date", "Charge"])

fig, ax = plt.subplots(figsize=(6, 4))
ax.axis("tight")
ax.axis("off")
table = plt.table(cellText=df.values, colLabels=df.columns, cellLoc="center", loc="center")
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width([0, 1])  

plt.savefig("/eos/user/s/shsong/www/dose/ME11_anode_charge.png")
plt.savefig("/eos/user/s/shsong/www/dose/ME11_anode_charge.pdf")
df.to_csv("/eos/user/s/shsong/www/dose/ME11_anode_charge.csv", index=False)

```
